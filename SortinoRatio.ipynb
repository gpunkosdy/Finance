```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
data = yf.download('AAPL', start='2023-1-1')['Adj Close']
data
```

```python
returns = pd.DataFrame(data.pct_change())
returns
```

```python
neg_returns = returns.loc[returns['Adj Close'] < 0]
neg_returns
```

```python
Sortino = (returns.mean()/neg_returns.std())*np.sqrt(252)
Sortino
```

```python
Sharpe = (returns.mean()/returns.std())*np.sqrt(252)
Sharpe
```
