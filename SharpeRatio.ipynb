```python
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
spy_price = yf.download('SPY', start='2021-1-1')['Adj Close']
```

```python
spy_price.info()
```

```python
returns = spy_price.pct_change()
print(returns)
print(returns.mean())
print(returns.std())
```

```python
sharpe = (returns.mean() / returns.std()) * np.sqrt(252)
sharpe
```

```python
spy_price.plot()
```

```python
np.cumprod(1 + returns).plot()
```
